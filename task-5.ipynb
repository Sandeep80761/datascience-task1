```python
df.target.value_counts()
```

```python
df.target.value_counts().plot(kind = 'bar',color=["#058577","#FDB409"])
plt.title("Heart diseases values")
plt.xlabel("1 = Heart Diseases, 0= No Heart Diseases")
plt.ylabel("No. of People")
plt.show()
```

```python
# Plotting a pie chart

df.target.value_counts().plot(kind = 'pie',figsize = (8,6))
plt.legend(["Heart Diseases", "Non Heart Diseases"],loc = 'upper right')
plt.show()
```

```python
# Sex Column....

# 0 represents `Female`

# 1 represents `Male`

# Target Column....

# 0 represents `No Heart Diseases`

# 1 represents `Heart Diseases`

# Let's check how many male and females are in dataset.

df.sex.value_counts()
```

```python
# Plotting a pie chart.

df.sex.value_counts().plot(kind='pie',autopct = '%1.2f%%')
plt.title("Sex Ratio Male and Female")
plt.show()
```

```python
pd.crosstab(df.sex,df.target)
```

```python
# SEX: 0 = Female , 1 = Male

sns.countplot(x='target',data=df,hue='sex')
plt.title("Heart Diseases frequency for sex")
plt.xlabel("0 = No Heart Diseases , 1 = Heart Diseases")
plt.show()
```

```python
# Total number of Female is 312 and Male is 713 which is double than female.

# 42.07% Male have Heart Diseases and 72.43% Female have Heart Diseases.
```

```python
df.cp.value_counts()
```

```python
df.cp.value_counts().plot(kind='bar',color = ["#058577","#FDB409","#2294F0","#5F7C8A"])
plt.title("Chest Pain type")
plt.show()
```

```python
# Comparing it with sex.

pd.crosstab(df.sex,df.cp)
```

```python
sns.countplot(x='sex',data=df,hue='cp',palette = 'rocket')
plt.title("Chest Pain type for sex")
plt.xlabel("0 = Female, 1 = Male")
plt.show()
```

```python
# Most of `Male` has `type 0` chest pain and least of `Male` has `type4` pain.

# But in case of `Female` `type 0` & `type 2` percentage has almost same.
```

```python
pd.crosstab(df.target,df.cp)
```

```python
ax = sns.countplot(x='cp',data=df,hue='target')
for label in ax.containers:
    ax.bar_label(label)
plt.show()
```

```python
# Most of people has `type 0` chest pain has less chance of Heart Diseases.

# And `type 2` chest pain has high chance of Heart Diseases.
```

```python
# 58-59 year old people are most in our dataset.

# Let's plot another distribution plot for `Maximum heart rate`

sns.displot(df['thalach'],bins=30,kde=True,color = 'chocolate')
```

```python
# From this plot we get a clear view about Maximum heart rate represented by `thalach`.
```

```python
sns.histplot(df['age'],bins=30,kde=True,color='#1A4870')
plt.title("Age Distribution by Patient")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize = (8,6))
ax = sns.countplot(x='thal',data=df,hue='target')
for label in ax.containers:
    ax.bar_label(label)
plt.title("Thalassemia Type vs. Heart Disease")
plt.xlabel("Thalassemia Type")
plt.ylabel("Count")
plt.legend(title='Heart Disease', loc='upper right')
plt.show()
```

```python
# Thalassemia Type 2 people has more chances of Heart Diseases than other types.

# In type 0 almost both ratio are almost same.

# Most of people has `type 3` Thalassemia has less chance of Heart Diseases.
```
